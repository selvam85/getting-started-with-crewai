# Kids content writer
This crew will collect stories from all cultures on different topics by researching the web and writing a compelling story. The purpose is to create a list of bed time stories that are interesting to kids.

```python
import os
from dotenv import load_dotenv

_ = load_dotenv()
os.environ["OPENAI_MODEL_NAME"] = "gpt-4-turbo"
```

```python
from crewai import Agent, Task, Crew
from crewai_tools import SerperDevTool, ScrapeWebsiteTool
```

## Outline

1. Research agent - Search the web for relevant stories and scrape the content
2. Story writer - writes the story based on the content from Research agent
3. Editor - Kids story editor who makes the story apt for kids
4. Publisher - publishes the story

## Create Agents

```python
research_agent = Agent(
    role="Content Researcher",
    goal="Research and find relevant content on {topic}",
    backstory="You are an expert at researching a topic in the internet "
        "and finding all the relevant content. You content that "
        "you research and collect will be used by authors to write stories "
        "apt for kids / family audiences. Your work is the basis for the "
        "Story Writer to write a story on this topic.",
    allow_delegation=False,
    verbose=True
)
```

```python
## Todo - once this works, we can make delegation true, to reach out to the content researcher asking for more relevant content, if necessary. 
story_writer_agent = Agent(
    role="Story Writer",
    goal="Write engaging stories for kids with beautiful narration",
    backstory="You are a talented and accomplished author who specializes on "
        "writing stories for kids. You have a special ability and taste in putting all "
        "the facts together in an engaging fashion for the kids.",
    allow_delegation=False,
    verbose=True
)
```

```python
##Todo - there is an opportunity to make delegation true, to send it back to the Story Writer to change the story, if necessary
editor_agent = Agent(
    role="Story Editor",
    goal="Edit a given story and ensure it is apt for kids to read",
    backstory="You are an accomplished editor and an expert in "
        "illustrating stories for kids. You have a talent to maximise "
        "the joy for the kids that read these stories. The output that "
        "you produce will be final version used for publishing.",
    allow_delegation=False,
    verbose=True
)
```

```python
##Todo - future stuff - ask the agent to publish the story in multiple language
publisher_agent = Agent(
    role="Content Publisher",
    goal="Publish the story in different formats",
    backstory="You are a content publisher who gives importance "
        "to the look and feel and format of the content. "
        "You have a lot of experience in kids book creation.",
    allow_delegation=False,
    verbose=True
)
```

## Create Tools

```python
search_tool = SerperDevTool()
website_scrape_tool = ScrapeWebsiteTool()
```

## Create Tasks

```python
content_research_task = Task(
    description=("1. Do a thorough research on the web "
                 "and collect all the relevant content on {topic}. "
                 "2. Ensure there is a lot of details so that the Story Writer can write "
                 "a very detailed story about all the events in great detail"
                 "3. Ensure that all the content gathered is from credible sources. "
                 "4. Do not base the content just on one source. "
                 "Always check on multiple sources so that all the details could be gathered. "
                 "5. Ensure that there is enough detail for the Story Writer to write "
                 "an engaging story with the content you gather."),
    expected_output="Detailed content on the given topic from credible sources in the web",
    agent=research_agent,
    tools=[search_tool, website_scrape_tool]
)
```

```python
story_writing_task = Task(
    description=("Review all the content produced by the Content Researcher "
                 "and write an engaging story for kids on {topic}. "
                 "The story should be a detailed one capturing all the events in detail "
                 "and it should atleast span for 1000 words or more. "
                 "Ensure you add context and background of the hero, the location, historical timeframe, etc. "),
    expected_output="Produce an engaging story for the kids based on the content "
        "received from Content Researcher",
    agent=story_writer_agent
)
```

```python
editing_task = Task(
    description=("Edit the story to make it more engaging and interesting "
                 "for kids. Ensure that the language is appropriate for kids "
                 "and correct all grammatical errors. Optimize for kids enjoyment "
                 "while they read the story."),
    expected_output="The final cut of the story that is ready for publishing",
    agent=editor_agent
)
```

```python
publishing_task = Task(
    description=("Publish the story received from Story Editor, formatted as markdown. "
                 "Generate a markdown file of the story."),
    expected_output="The story is published as markdown and a markdown file generated",
    output_file="story_1.md",
    agent=publisher_agent
)
```

## Create Crew

```python
story_publishing_crew = Crew(
    agents=[research_agent,
            story_writer_agent,
            editor_agent,
            publisher_agent],
    tasks=[content_research_task,
           story_writing_task,
           editing_task,
           publishing_task],
    verbose=True
)
```

```python
result = story_publishing_crew.kickoff(inputs={"topic": "Chatrapathi Shivaji using tiger claws"})
```
